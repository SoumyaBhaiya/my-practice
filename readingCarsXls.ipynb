```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
cars = pd.read_excel("cars.xls")
cars.head() 
```

```python
import statsmodels.api as sm 
```

```python
cars['Model_ord'] = pd.Categorical(cars.Model).codes
```

```python
cars.head()
```

```python
X = cars[['Mileage', 'Model_ord', 'Doors']]
y = cars[['Price']]
```

```python
X1 = sm.add_constant(X)
```

```python
X1
```

```python
est = sm.OLS(y, X1).fit()
```

```python
est
```

```python
est.summary()
```

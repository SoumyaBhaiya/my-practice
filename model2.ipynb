```python
def remove_outliers(data):
    median = np.median(data)
    std = np.std(data)
    filter_values =[]
    for e in data:
        if median - 2*std < e < median + 2*std:
            filter_values.append(e)
    return np.array(filter_values)

```

```python
filtered = remove_outliers(incomes)
plt.hist(filtered, 50)
plt.show()
```

```python
#Generating another of the sample data.

#Gonna use conditional proabability on this fake data. the older people gonna be more likely to buy something and the younger one will be less likely
```

```python
np.random.seed(0)
```

```python
#making dictionaries to keep count. numbers gonna be the ages, and the rest would be the number of buy/values
```

```python
totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchasing = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0
```

```python
for _ in range(1000000):
    ageDecade = np.random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = float(ageDecade)/100.00
    totals[ageDecade] +=1 
    if (np.random.random() < purchaseProbability):
        totalPurchases +=1
        purchasing[ageDecade] += 1
```

```python
print(totals)
print(purchasing)
print(purchaseProbability)
```

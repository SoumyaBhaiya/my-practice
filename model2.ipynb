```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
incomes = np.random.normal(10000, 7500, 10000)
```

```python
incomes = np.append(incomes, [ 10000000])
```

```python
incomes
```

```python
incomes.mean()
```

```python
plt.hist(incomes, 50)
plt.show()
```

```python
def remove_outliers(data):
    median = np.median(data)
    std = np.std(data)
    filter_values =[]
    for e in data:
        if median - 2*std < e < median + 2*std:
            filter_values.append(e)
    return np.array(filter_values)

```

```python
filtered = remove_outliers(incomes)
plt.hist(filtered, 50)
plt.show()
```
